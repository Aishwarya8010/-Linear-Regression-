```python
###### Importing of pacakage----
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('insurance.csv')
```

```python
print(data.head())
```

```python
#### We have predict the insurance based on charges -
df = data[['charges','bmi']]
df.head(2)
```

```python
df.shape
```

```python
#### Sampling of data ---
training,testing = train_test_split(data,test_size=0.2)
```

```python
training.head(5)
```

```python
testing.shape
```

```python
#### Defining------
x = training['charges']
y = training['bmi']
```

```python
#### Only for Linear Regression it takes an array like shape with one dimension only.
x = x.values.reshape(-1,1)
y = y.values.reshape(-1,1)
```

```python
#### linear Regression Linear model-----
from sklearn.linear_model import LinearRegression
```

```python
##### We have to create a object of this linear Regression model----
model = LinearRegression()              ### y = mx + b
print(model)
```

```python
train_model = model.fit(x,y)
print("Model training is Done")
print(train_model)
```

```python
print(model.coef_)         #### this is a m value
print(model.intercept_)    ###3 y = mx = b this is b
```

```python
##### Equation after training model 
x1 = 27900
y = 8.88575721e-05*x1 + 29.51455427
print("The bmi of {} is {}".format(x1,y))
```

```python
##### For validation we are passing only x and it is predicting y 
x_test = testing['bmi'].values.reshape(-1,1)
x_test
```

```python
pred_charges = train_model.predict(x_test)
pred_charges
```

```python
testing['pred_charges'] = pred_charges
testing
```

```python
testing['Error'] = testing['charges'] - testing['pred_charges']       ###### actual - predicted
testing
```

```python
sns.regplot(x = 'bmi',y = 'pred_charges',data = testing)
```

```python
sns.regplot(x = 'bmi',y = 'charges',data = testing)
```
